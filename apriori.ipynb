```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

from apyori import apriori
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
df2 = pd.read_csv('/content/drive/MyDrive/TA BARU/datasetnya/mhs2020coba.csv', sep = ';', index_col = 0)
```

```python
df2.info(verbose=True)
```

```python
frequent_itemsets = apriori(df2, min_support=0.15, use_colnames=True)
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold = 1)
rules
```

Grafik Support dan Confidence seluruh aturan

```python
import matplotlib.pyplot as plt

# Ambil nilai support dan confidence dari DataFrame rules
support = rules['support'].values.flatten()
confidence = rules['confidence'].values.flatten()
rules_list = rules.index.astype(str)  # Ubah indeks menjadi string untuk digunakan sebagai label sumbu x

# Plot line chart untuk support dan confidence
plt.figure(figsize=(12, 6))
plt.plot(rules_list, support, marker='o', color='blue', label='Support')
plt.plot(rules_list, confidence, marker='o', color='red', label='Confidence')
plt.xlabel('Rules')
plt.ylabel('Metrics')
plt.title('Support dan Confidence untuk Setiap Aturan')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)  # Tambahkan grid
plt.tight_layout()  # Rapihkan tata letak
plt.show()
```

FINDING BOND MEASUREMENT

```python
frequent_itemsets.reset_index(drop=True, inplace=True)
itemsets_array = frequent_itemsets['itemsets'].iloc[92]
itemsets_array
```

```python
result = frequent_itemsets[frequent_itemsets['itemsets'].apply(
    lambda x: any(item in x for item in itemsets_array))]
result
```

```python
target_itemset = {'A.1FUNDIF', 'A.3MATLAN'}

# Find the row corresponding to the target itemset
target_row = frequent_itemsets[frequent_itemsets['itemsets'] == target_itemset]

# Retrieve the support value
support_value = target_row['support'].values[0]

print(f"Support for itemset {target_itemset}: {support_value}")
```

```python
#Conjunctive Sup
conjunctive_items = {'A.1FUNDIF', 'A.3MATLAN'}

matching_itemsets = frequent_itemsets['itemsets'][frequent_itemsets['itemsets'].apply(
    lambda x: all(item in x for item in conjunctive_items))].tolist()

# Display the list of itemsets
print(f"Itemsets containing the conjunctive items {conjunctive_items}:")
for itemset in matching_itemsets:
    print(itemset)

# Count the number of itemsets containing all conjunctive items
num_conjunctive_occurrences = frequent_itemsets['itemsets'].apply(
    lambda x: all(item in x for item in conjunctive_items)).sum()

# Display the count
print(f"Number of itemsets containing the conjunctive items: {num_conjunctive_occurrences}")
```

```python
#Disjunctive Sup
disjunctive_items = {'A.1FUNDIF', 'A.3MATLAN'}

# Find the rows corresponding to itemsets containing at least one disjunctive item
matching_rows = frequent_itemsets[frequent_itemsets['itemsets'].apply(
    lambda x: any(item in x for item in disjunctive_items))]

# Retrieve the disjunctive support values
disjunctive_support_values = matching_rows['support']

# Display the disjunctive support values
print("Disjunctive Support Values:")
for index, support_value in zip(matching_rows.index, disjunctive_support_values):
    print(f"Itemset {frequent_itemsets['itemsets'][index]}: {support_value}")

#Count the number of itemsets containing at least one disjunctive item
num_disjunctive_occurrences = frequent_itemsets['itemsets'].apply(
    lambda x: any(item in x for item in disjunctive_items)).sum()

print(f"Number of itemsets containing the disjunctive items: {num_disjunctive_occurrences}")
```

```python
# Calculate the bond
bond_value = num_conjunctive_occurrences / num_disjunctive_occurrences if num_disjunctive_occurrences != 0 else 0
print(f"Bond value: {bond_value}")
```
